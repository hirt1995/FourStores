# 1. Importación de datos


```python
import pandas as pd

#Leemos los archivos del repositorio
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

#Cargamos y etiquetamos los datos obtenidos de los archivos csv. Identificamos el origen: agregamos una columna "Tienda" que indique de qué archivo proviene cada registro.
tienda1 = pd.read_csv(url);  tienda1["Tienda"] = "Tienda 1"
tienda2 = pd.read_csv(url2); tienda2["Tienda"] = "Tienda 2"
tienda3 = pd.read_csv(url3); tienda3["Tienda"] = "Tienda 3"
tienda4 = pd.read_csv(url4); tienda4["Tienda"] = "Tienda 4"

#Unimos los datos de las tiendas
tiendas = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)

#convertimos las fechas para su uso como datos
tiendas["Fecha de Compra"] = pd.to_datetime(tiendas["Fecha de Compra"], dayfirst=True)

#Verificamos la estructura de neustro nuevo dataframe
tiendas.info()
tiendas.head()
```

# 6. Productos más y menos vendidos

#### **Categoría de productos más vendidos y su valor en COP/USD por cada tienda**

```python
#¿cuál es al categoria de productos que genera más ganancias a EN PESOS Y DÓLARES y el segundo producto que más genera para cada tienda?
GanPorTienda_numeric = tiendas.groupby(["Tienda", "Categoría del Producto"])["Precio"].sum().reset_index()
TASA_USD = 3865 # valor de tasa de cambio de dolar fijada

# Aplicamos nlargest al valor numérico de la columna 'Precio'
top_earning_categories = GanPorTienda_numeric.groupby("Tienda")["Precio"].nlargest(2).reset_index()

# Combinamos GanPorTienda_numeric  para obetener el nombre de la categoria
top_earning_categories_formatted = top_earning_categories.merge(GanPorTienda_numeric[['Tienda', 'Categoría del Producto', 'Precio']],on=['Tienda', 'Precio'])

# Formatemos para observan mejor en pantalla en pesos colombianos y dólares
top_earning_categories_formatted['Precio_COP'] = top_earning_categories_formatted["Precio"].apply(lambda x: f"${x:,.0f} COP")
top_earning_categories_formatted['Precio_USD'] = top_earning_categories_formatted["Precio"].apply(lambda x: f"${x/TASA_USD:,.2f} USD")

# Quitamos la columna precio 'Precio' para visualización
top_earning_categories_formatted.drop('Precio', axis=1, inplace=True)
top_earning_categories_formatted.drop('level_1', axis=1, inplace=True)

top_earning_categories_formatted
```

#### **Categoría de productos menos vendidos y su valor en COP/USD por cada tienda**

```python
#¿cuál es al categoria de productos que genera menos ganancias a EN PESOS Y DÓLARES y el segundo producto que más genera para cada tienda?
GanPorTienda_numeric = tiendas.groupby(["Tienda", "Categoría del Producto"])["Precio"].sum().reset_index()
TASA_USD = 3865 # valor de tasa de cambio de dolar fijada

# Aplicamos nlargest al valor numérico de la columna 'Precio'
low_earning_categories = GanPorTienda_numeric.groupby("Tienda")["Precio"].nsmallest(2).reset_index()

# Combinamos GanPorTienda_numeric  para obetener el nombre de la categoria
low_earning_categories_formatted = low_earning_categories.merge(GanPorTienda_numeric[['Tienda', 'Categoría del Producto', 'Precio']],on=['Tienda', 'Precio'])

# Formatemos para observan mejor en pantalla en pesos colombianos y dólares
low_earning_categories_formatted['Precio_COP'] = low_earning_categories_formatted["Precio"].apply(lambda x: f"${x:,.0f} COP")
low_earning_categories_formatted['Precio_USD'] = low_earning_categories_formatted["Precio"].apply(lambda x: f"${x/TASA_USD:,.2f} USD")

# Quitamos la columna precio 'Precio' para visualización
low_earning_categories_formatted.drop('Precio', axis=1, inplace=True)
low_earning_categories_formatted.drop('level_1', axis=1, inplace=True)

low_earning_categories_formatted
```

#### **Productos más y menos vendidos vendidos y su valor en COP/USD por cada tienda**

```python
#Queremos saber los pruductos más vendidos y menos vendidos por cada tienda con su valor en COP y USD
ganancia_por_producto = tiendas.groupby(["Tienda", "Producto"])["Precio"].sum()
TASA_USD = 3865 # Assuming this is already defined

# Obtenemos un productos de cada tienda que sea el más vendido
top_products = ganancia_por_producto.groupby("Tienda").nlargest(1).reset_index(level=1)
top_products.columns = ['Producto', 'Precio']
top_products["Precio_COP"] = top_products["Precio"].apply(lambda x: f"${x:,.0f} COP")
top_products["Precio_USD"] = top_products["Precio"].apply(lambda x: f"${x/TASA_USD:,.2f} USD")
top_products.drop("Precio", axis=1, inplace=True)
print("Top productos mas vendidos por tienda:")
display(top_products)

# Obtenemos un productos de cada tienda que sea el menos vendido
bottom_products = ganancia_por_producto.groupby("Tienda").nsmallest(1).reset_index(level=1)
bottom_products.columns = ['Producto', 'Precio']
bottom_products["Precio_COP"] = bottom_products["Precio"].apply(lambda x: f"${x:,.0f} COP")
bottom_products["Precio_USD"] = bottom_products["Precio"].apply(lambda x: f"${x/TASA_USD:,.2f} USD")
bottom_products.drop("Precio", axis=1, inplace=True)
print("\nTop productos menos vendidos por tienda:")
display(bottom_products)
```

#### **Vendedor con mayor volumen de articulos vendidos, mejor calificado, peor calificado por tienda**

```python
# Queremos ver el vendedor con mayor volumen de articulos vendidos por tienda,
# el de mejor calificación, y el de peor calificación.

#Calculamos el el total de artículos vendidos y el promedio de calificacion para cada vendedor por tienda
vendor_sales_summary = tiendas.groupby(["Tienda", "Vendedor"]).agg(
    Total_Articulos_Vendidos=('Producto', 'count'),
    Promedio_Calificación=('Calificación', 'mean')
).reset_index()

#Encotramos el vendedor con el mayor volumen de articulos vendidos en cada tienda
top_volume_vendor = vendor_sales_summary.loc[vendor_sales_summary.groupby('Tienda')['Total_Articulos_Vendidos'].idxmax()]
print("El vendedor con la mayor cantidad de articulos vendidos:")
display(top_volume_vendor)

# Encotnramos el vendedor con la mejor calificación
best_rated_vendor = vendor_sales_summary.loc[vendor_sales_summary.groupby('Tienda')['Promedio_Calificación'].idxmax()]
print("\nVendedor con la mejor calificación por tienda:")
display(best_rated_vendor)

# Encotnramos el vendedor peor calificado por tienda
worst_rated_vendor = vendor_sales_summary.loc[vendor_sales_summary.groupby('Tienda')['Promedio_Calificación'].idxmin()]
print("\nVendedor peor calificado por tienda:")
display(worst_rated_vendor)
```

Implementar planes de capacitación o mejora del rendimiento dirigidos a los vendedores con un alto volumen de ventas pero baja satisfacción del cliente, ya que su impacto negativo en la experiencia del cliente podría superar sus contribuciones de ventas.

```python
#Calculamos el total de ventas y calificación para cada vendedor a trvés de todas las tiendas y sacamos su rendimiento
vendor_performance = tiendas.groupby('Vendedor').agg(
    Total_Sales=('Producto', 'count'),
    Numero_Calificaciones_Superiores_a_4=('Calificación', lambda x: (x > 4).sum()),
    Numero_Calificaciones_Inferiores_a_3=('Calificación', lambda x: (x < 3).sum())
).reset_index()

# Calculamos los porcentajes
vendor_performance['% Calificaciones Superiores a 4'] = (vendor_performance['Numero_Calificaciones_Superiores_a_4'] / vendor_performance['Total_Sales']) * 100
vendor_performance['% Calificaciones Inferiores a 3'] = (vendor_performance['Numero_Calificaciones_Inferiores_a_3'] / vendor_performance['Total_Sales']) * 100
vendor_performance['% Otras Calificaciones'] = 100 - (vendor_performance['% Calificaciones Superiores a 4'] + vendor_performance['% Calificaciones Inferiores a 3'])

# Calculamos la calificación promedio para cada vendedor en cada tienda
vendor_store_ratings = tiendas.groupby(['Vendedor', 'Tienda'])['Calificación'].mean().reset_index(name='Average_Rating')

# Encontramos la tienda en donde cada vendedor tuvo mejor calificación
best_store_rating_idx = vendor_store_ratings.groupby('Vendedor')['Average_Rating'].idxmax()
best_store_rating = vendor_store_ratings.loc[best_store_rating_idx][['Vendedor', 'Tienda', 'Average_Rating']].rename(
    columns={'Tienda': 'Best_Performance_Store', 'Average_Rating': 'Best_Average_Rating'}
)

# Encontramos la tienda donde cada vendedor obtuvo la peor calificacion
worst_store_rating_idx = vendor_store_ratings.groupby('Vendedor')['Average_Rating'].idxmin()
worst_store_rating = vendor_store_ratings.loc[worst_store_rating_idx][['Vendedor', 'Tienda', 'Average_Rating']].rename(
    columns={'Tienda': 'Worst_Performance_Store', 'Average_Rating': 'Worst_Average_Rating'}
)

# Combinamos toda la información mejores y peores
best_worst_store_rating = best_store_rating.merge(worst_store_rating, on='Vendedor')


# Combinamos los datos de redimiento con los datos mejores/peores
final_vendor_ranking = vendor_performance.merge(best_worst_store_rating, on='Vendedor', how='left')

# Rankeamos los vendedores basandonos en el rendimiendo (tenidendo en cuenta diferencia entre procentajes positivos y negativos)
final_vendor_ranking['Performance_Score'] = final_vendor_ranking['% Calificaciones Superiores a 4'] - final_vendor_ranking['% Calificaciones Inferiores a 3']

# Organizamos de forma descendente por el score de rendimiento
final_vendor_ranking = final_vendor_ranking.sort_values(by='Performance_Score', ascending=False)

# Formateamos columnas a 2 decimales
final_vendor_ranking['% Calificaciones Superiores a 4'] = final_vendor_ranking['% Calificaciones Superiores a 4'].map('{:.2f}'.format)
final_vendor_ranking['% Calificaciones Inferiores a 3'] = final_vendor_ranking['% Calificaciones Inferiores a 3'].map('{:.2f}'.format)
final_vendor_ranking['% Otras Calificaciones'] = final_vendor_ranking['% Otras Calificaciones'].map('{:.2f}'.format)


# Mostramos con las calificaciones y procentajes de rendimiento
print("Vendor Ranking by Performance:")
display(final_vendor_ranking[['Vendedor', 'Total_Sales', 'Numero_Calificaciones_Superiores_a_4', 'Numero_Calificaciones_Inferiores_a_3', '% Calificaciones Superiores a 4', '% Calificaciones Inferiores a 3', '% Otras Calificaciones', 'Best_Performance_Store', 'Worst_Performance_Store', 'Performance_Score']])
```
